```python
#creamos una clase
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
#una funcion que sera metodo
    def mostrar(self):
        print(f'Nombre: {self.nombre}, Edad: {self.edad}')


persona1 = Persona('Juan', 22)
persona1.mostrar()

persona2 = Persona('Maria', 30)
persona2.mostrar()
```
